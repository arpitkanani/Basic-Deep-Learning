```python
import tensorflow as tf
import os
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import time
import seaborn as sns
plt.style.use("fivethirtyeight")
%load_ext tensorboard
```

```python
mnist=tf.keras.datasets.mnist
```

```python
(X_train_full,y_train_full),(X_test,y_test)=mnist.load_data()
X_valid,X_train=X_train_full[:5000]/255.,X_train_full[5000:]/255.
y_valid,y_train=y_train_full[:5000],y_train_full[5000:]

X_test=X_test/255.
```

```python
X_train
```

```python
from tensorflow.keras import regularizers
from tensorflow.keras.layers import Dropout

tf.random.set_seed(42)
np.random.seed(42)

##CReating layers of ANN

Layers=[tf.keras.layers.Flatten(input_shape=[28,28],name='inputLayer'),
        tf.keras.layers.Dense(300,activation='relu',name='hiddenLayer1',kernel_regularizer=regularizers.L2(1e-4),kernel_initializer='he_normal'),
        tf.keras.layers.LeakyReLU(),
        #tf.keras.layers.BatchNormalization(),
        tf.keras.layers.Dense(100,activation='relu',name='hiddenLayer2'),
        tf.keras.layers.LeakyReLU(),
        #tf.keras.layers.Dropout(0.2),
        tf.keras.layers.Dense(10,activation='softmax',name='outputLayer')]

model=tf.keras.models.Sequential(Layers)
```

```python
loss_fn='sparse_categorical_crossentropy'
optimizer='SGD'
metrics=['accuracy']

model.compile(loss=loss_fn,
              optimizer=optimizer,
              metrics=metrics
                )
```

```python
start=time.time()

epochs=10
validation_set=(X_valid,y_valid)

history=model.fit(X_train,y_train,epochs=epochs,validation_data=validation_set,verbose=2)

end=time.time()

print(f"Runtime of Program is {end-start}")
```

```python
Layers_BN=[
    tf.keras.layers.Flatten(input_shape=[28,28],name='inputLayer'),
    tf.keras.layers.BatchNormalization(),
    tf.keras.layers.Dense(300,activation='relu',name='hiddenLayer1',kernel_regularizer=regularizers.L2(1e-4),kernel_initializer='he_normal'),
    tf.keras.layers.LeakyReLU(),
    tf.keras.layers.BatchNormalization(),
    tf.keras.layers.Dense(100,activation='relu',name='hiddenLayer2'),
    tf.keras.layers.LeakyReLU(),
    tf.keras.layers.BatchNormalization(),
    tf.keras.layers.Dropout(0.2),
    tf.keras.layers.Dense(10,activation='softmax',name='outputLayer')
]

model=tf.keras.models.Sequential(Layers_BN)
```
