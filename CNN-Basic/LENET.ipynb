```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns 
import tensorflow as tf
from tensorflow import keras
import keras
import numpy as np
from keras.models import Sequential
from keras.layers import Dropout,Flatten,Conv2D,MaxPooling2D,Dense,GlobalAveragePooling2D,BatchNormalization
from keras.regularizers import Regularizer
from tensorflow.keras.datasets import mnist
from tensorflow.keras.callbacks import ModelCheckpoint
```

```python
(X_train,y_train),(X_test,y_test)=mnist.load_data()
X_train.shape,y_train.shape
```

```python
X_valid,X_train=X_train[:5000]/255. ,X_train[5000:]/255.
y_valid,y_train=y_train[:5000],y_train[5000:] 
X_test=X_test/255.
```

```python
img_rows,img_cols=28,28
X_train=X_train.reshape(X_train.shape[0],img_rows,img_cols,1)
X_test=X_test.reshape(X_test.shape[0],img_rows,img_cols,1)

input_shape=(img_rows,img_cols,1)

input_shape,X_train.shape

```

```python
y_test[:5]
```

```python
from tensorflow.keras.utils import to_categorical
y_train=keras.utils.to_categorical(y_train,10)
y_test=keras.utils.to_categorical(y_test,10)
```

```python
model=Sequential()
```

#### in batch normaization 4 parameter are there in that 2 are trainable(alph and beta) and 2 are non trainable (SD,mean)
it calculate by channel_size*4

```python

#CONV_1 
model.add(Conv2D(6,kernel_size=(5,5),activation='relu',input_shape=(28,28,1)))

#POOL_1
model.add(MaxPooling2D(pool_size=(2,2)))

#CONVO_2
model.add(Conv2D(16,(5,5),activation='relu'))

#POOl_2
model.add(MaxPooling2D(pool_size=(2,2)))

#Flatten here too many dimension so we want only classifiaction o/p
model.add(Flatten())

#FC_1
model.add(Dense(120,activation='relu'))

#model.add(BatchNormalization())
model.add(Dense(84,activation='relu'))
#model.add(Dropout(0.2))

#FC_2
model.add(Dense(10,activation='softmax'))
model.summary()

model.compile(loss='categorical_crossentropy',optimizer='adam',metrics=['accuracy'])

history=model.fit(X_train,y_train,batch_size=128,epochs=2,validation_data=(X_test,y_test)) # type: ignore
score=model.evaluate(X_test,y_test)
score[0],score[1]

```
